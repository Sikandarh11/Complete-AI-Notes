# 1. TWO SUM PROBLEM

```python
"""Problem: Find two numbers in array that sum to target
Input: nums = [2, 7, 11, 15], target = 9
Output: [0, 1] (indices of 2 and 7)
"""

# APPROACH 1: Hash Map (BEST - O(n) time, O(n) space)
def twoSum(nums, target):
    seen = {}  # num -> index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []  # No solution

# Test
print(twoSum([2, 7, 11, 15], 9))  # [0, 1]

# How it works:
# nums = [2, 7, 11, 15], target = 9
# i=0, num=2: complement=7, not in seen. Add 2->0 to seen
# i=1, num=7: complement=2, found in seen! Return [seen[2], 1] = [0, 1]

# APPROACH 2: Brute Force (O(n²) time, O(1) space) - AVOID
def twoSum_bruteforce(nums, target):
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []

# APPROACH 3: Two Pointers (if sorted) - O(n) time, O(1) space
def twoSum_sorted(nums, target):
    sorted_nums = sorted(enumerate(nums), key=lambda x: x[1])
    left, right = 0, len(sorted_nums) - 1
    while left < right:
        curr_sum = sorted_nums[left][1] + sorted_nums[right][1]
        if curr_sum == target:
            return [sorted_nums[left][0], sorted_nums[right][0]]
        elif curr_sum < target:
            left += 1
        else:
            right -= 1
    return []

# Interview tip: Use hash map approach. It's fast, clean, and shows you understand trade-offs
```

# 2. VALID PARENTHESES

```python
"""Problem: Check if parentheses are balanced
Input: '()[]{}' -> True
Input: '([)]' -> False (wrong order)
Input: '()' -> True
"""

# STACK APPROACH (BEST - O(n) time, O(n) space)
def isValid(s):
    """Check if parentheses are valid and balanced."""
    stack = []
    pairs = {'(': ')', '{': '}', '[': ']'}
    
    for char in s:
        if char in pairs:  # Opening bracket
            stack.append(char)
        else:  # Closing bracket
            if not stack or pairs[stack.pop()] != char:
                return False
    
    return len(stack) == 0  # All opened brackets must be closed

# Test cases
print(isValid('()[]{}'))  # True
print(isValid('([)]'))    # False - wrong nesting
print(isValid('{[]}'))    # True
print(isValid(''))        # True - empty is valid
print(isValid('('))       # False - unclosed

# How it works for '({[]})':
# char='(': opening, stack=['(']
# char='{': opening, stack=['(', '{']
# char='[': opening, stack=['(', '{', '[']
# char=']': closing, pop '[', match? Yes! stack=['(', '{']
# char='}': closing, pop '{', match? Yes! stack=['(']
# char=')': closing, pop '(', match? Yes! stack=[]
# stack empty? Yes -> True

# Interview tip: 
# - Mention stack is ideal for matching brackets
# - Edge cases: empty string, single bracket, mismatched pairs
```

# 3. BINARY SEARCH

```python
"""Problem: Find target in SORTED array
Input: nums = [1, 3, 5, 7, 9], target = 5
Output: 2 (index)
"""

# ITERATIVE (BEST - O(log n) time, O(1) space)
def binarySearch(nums, target):
    """Iterative binary search."""
    left, right = 0, len(nums) - 1
    
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1  # Search right half
        else:
            right = mid - 1  # Search left half
    
    return -1  # Not found

# Test
print(binarySearch([1, 3, 5, 7, 9], 5))    # 2
print(binarySearch([1, 3, 5, 7, 9], 1))    # 0
print(binarySearch([1, 3, 5, 7, 9], 10))   # -1

# Trace for binarySearch([1, 3, 5, 7, 9], 5):
# left=0, right=4: mid=2, nums[2]=5, target=5 -> FOUND! return 2

# RECURSIVE
def binarySearch_recursive(nums, target, left=0, right=None):
    """Recursive binary search."""
    if right is None:
        right = len(nums) - 1
    
    if left > right:
        return -1
    
    mid = (left + right) // 2
    if nums[mid] == target:
        return mid
    elif nums[mid] < target:
        return binarySearch_recursive(nums, target, mid + 1, right)
    else:
        return binarySearch_recursive(nums, target, left, mid - 1)

# Interview tip:
# - Must be sorted!
# - Use left <= right (not left < right) - subtle bug
# - mid = left + (right - left) // 2 avoids overflow in other languages
# - Variations: find first/last occurrence, rotated array
```

# 4. LONGEST SUBSTRING WITHOUT REPEATING CHARACTERS

```python
"""Problem: Find length of longest substring without repeating chars
Input: 'abcabcbb'
Output: 3 ('abc')
Input: 'dvdf'
Output: 3 ('vdf')
"""

# SLIDING WINDOW + HASH MAP (BEST - O(n) time, O(min(n, charset)) space)
def lengthOfLongestSubstring(s):
    """Find length of longest substring without repeating chars."""
    char_index = {}  # char -> last seen index
    max_length = 0
    start = 0  # Window start
    
    for end in range(len(s)):
        char = s[end]
        
        # If char seen and within current window, move start
        if char in char_index and char_index[char] >= start:
            start = char_index[char] + 1
        
        char_index[char] = end  # Update last seen index
        max_length = max(max_length, end - start + 1)
    
    return max_length

# Test
print(lengthOfLongestSubstring('abcabcbb'))  # 3 ('abc')
print(lengthOfLongestSubstring('bbbbb'))     # 1 ('b')
print(lengthOfLongestSubstring('au'))        # 2 ('au')
print(lengthOfLongestSubstring('dvdf'))      # 3 ('vdf')

# Trace for 'dvdf':
# start=0, max_length=0
# end=0, char='d': char_index={'d':0}, max_len=1
# end=1, char='v': char_index={'d':0, 'v':1}, max_len=2
# end=2, char='d': 'd' in dict, index 0 >= start 0, start=1. char_index={'d':2, 'v':1}, max_len=2
# end=3, char='f': char_index={'d':2, 'v':1, 'f':3}, max_len=max(2, 3-1+1)=3
# Return 3

# Interview tip:
# - Sliding window is KEY pattern
# - Two pointers (start, end) expand/contract window
# - Hash map tracks last seen index (efficient lookup)
# - Condition: char_index[char] >= start (char within window)
```

# 5. MERGE SORTED ARRAYS

```python
"""Problem: Merge two sorted arrays
Input: arr1 = [1,2,3], arr2 = [2,5,6]
Output: [1,2,2,3,5,6]
"""

# APPROACH 1: Two Pointers (BEST - O(n+m) time, O(n+m) space)
def merge(arr1, arr2):
    """Merge two sorted arrays."""
    result = []
    i, j = 0, 0
    
    # Compare and add smaller element
    while i < len(arr1) and j < len(arr2):
        if arr1[i] <= arr2[j]:
            result.append(arr1[i])
            i += 1
        else:
            result.append(arr2[j])
            j += 1
    
    # Add remaining elements
    result.extend(arr1[i:])
    result.extend(arr2[j:])
    
    return result

# Test
print(merge([1, 2, 3], [2, 5, 6]))  # [1, 2, 2, 3, 5, 6]
print(merge([1], [2]))              # [1, 2]
print(merge([], [1, 2]))            # [1, 2]

# APPROACH 2: In-place (O(n+m) time, O(1) space) - if space allowed in arr1
def merge_inplace(arr1, m, arr2, n):
    """Merge arr2 into arr1 (arr1 has extra space).
    arr1 = [1,2,3,0,0,0], m=3, arr2 = [2,5,6], n=3
    Result: arr1 becomes [1,2,2,3,5,6]
    """
    p1, p2, p = m - 1, n - 1, m + n - 1  # Work backwards
    
    while p1 >= 0 and p2 >= 0:
        if arr1[p1] > arr2[p2]:
            arr1[p] = arr1[p1]
            p1 -= 1
        else:
            arr1[p] = arr2[p2]
            p2 -= 1
        p -= 1
    
    # If arr2 has remaining, copy them (arr1 remaining already in place)
    arr1[:p2+1] = arr2[:p2+1]

# Interview tip:
# - Standard: two pointers from start, build result
# - In-place: two pointers from END (avoid overwriting)
# - Know both approaches, discuss space trade-off
```

# 6. REVERSE A LIST / STRING

```python
"""Problem: Reverse a list/string
Input: [1, 2, 3, 4, 5]
Output: [5, 4, 3, 2, 1]
"""

# APPROACH 1: Two Pointers In-place (BEST - O(n) time, O(1) space)
def reverse(arr):
    """Reverse array in-place."""
    left, right = 0, len(arr) - 1
    
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]  # Swap
        left += 1
        right -= 1
    
    return arr

# Test
print(reverse([1, 2, 3, 4, 5]))  # [5, 4, 3, 2, 1]

# APPROACH 2: Pythonic Slice (O(n) time, O(n) space)
arr = [1, 2, 3, 4, 5]
print(arr[::-1])  # [5, 4, 3, 2, 1] - Creates new list

# APPROACH 3: Recursive (O(n) time, O(n) space for call stack)
def reverse_recursive(arr, left=0, right=None):
    """Reverse array recursively."""
    if right is None:
        right = len(arr) - 1
    
    if left >= right:
        return arr
    
    arr[left], arr[right] = arr[right], arr[left]
    return reverse_recursive(arr, left + 1, right - 1)

print(reverse_recursive([1, 2, 3, 4, 5]))  # [5, 4, 3, 2, 1]

# REVERSE STRING
def reverse_string(s):
    """Reverse string (convert to list, reverse, rejoin)."""
    chars = list(s)  # String is immutable, convert to list
    left, right = 0, len(chars) - 1
    
    while left < right:
        chars[left], chars[right] = chars[right], chars[left]
        left += 1
        right -= 1
    
    return ''.join(chars)

print(reverse_string('hello'))  # 'olleh'

# Interview tip:
# - In-place is O(1) space (preferred in interviews)
# - Slice notation is Pythonic but uses extra space
# - Show both iterative and recursive approaches
```

# 7. FIBONACCI SEQUENCE

```python
"""Problem: Calculate nth Fibonacci number
0, 1, 1, 2, 3, 5, 8, 13...
fib(5) = 5
"""

# APPROACH 1: Memoization (Dynamic Programming) (BEST - O(n) time, O(n) space)
def fib_memo(n, memo={}):
    """Fibonacci with memoization (cache results)."""
    if n in memo:
        return memo[n]
    
    if n <= 1:
        return n
    
    memo[n] = fib_memo(n-1, memo) + fib_memo(n-2, memo)
    return memo[n]

# Or use @functools.lru_cache (cleaner)
from functools import lru_cache

@lru_cache(maxsize=None)
def fib_cached(n):
    """Fibonacci with @lru_cache decorator."""
    if n <= 1:
        return n
    return fib_cached(n-1) + fib_cached(n-2)

print(fib_cached(10))  # 55

# APPROACH 2: Tabulation (Bottom-up DP) (O(n) time, O(n) space)
def fib_tab(n):
    """Fibonacci with tabulation (build table from bottom)."""
    if n <= 1:
        return n
    
    dp = [0] * (n + 1)
    dp[1] = 1
    
    for i in range(2, n + 1):
        dp[i] = dp[i-1] + dp[i-2]
    
    return dp[n]

print(fib_tab(10))  # 55

# APPROACH 3: Space-optimized (O(n) time, O(1) space)
def fib_optimized(n):
    """Fibonacci with constant space (only track last 2)."""
    if n <= 1:
        return n
    
    prev, curr = 0, 1
    for _ in range(2, n + 1):
        prev, curr = curr, prev + curr
    
    return curr

print(fib_optimized(10))  # 55

# APPROACH 4: Naive Recursion (AVOID - O(2^n) time)
def fib_naive(n):
    """Naive recursive - exponential time!"""
    if n <= 1:
        return n
    return fib_naive(n-1) + fib_naive(n-2)

# fib_naive(40) will take FOREVER (recalculates same values)

# Interview tip:
# - Naive recursion has overlapping subproblems
# - Memoization: top-down DP (recursion + cache)
# - Tabulation: bottom-up DP (no recursion)
# - Optimize space when possible
```

# 8. CLIMBING STAIRS (DP VARIANT)

```python
"""Problem: You can climb 1 or 2 stairs at a time. How many ways to reach n?
n=1: 1 way (1)
n=2: 2 ways (1+1, 2)
n=3: 3 ways (1+1+1, 1+2, 2+1)
n=4: 5 ways (1+1+1+1, 1+1+2, 1+2+1, 2+1+1, 2+2)
Pattern: Same as Fibonacci!
"""

# Dynamic Programming Solution
def climbStairs(n):
    """Count ways to climb n stairs (1 or 2 at a time)."""
    if n <= 2:
        return n
    
    # dp[i] = ways to reach stair i
    # dp[i] = dp[i-1] + dp[i-2] (from stair i-1 or i-2)
    prev2, prev1 = 1, 2
    
    for i in range(3, n + 1):
        curr = prev1 + prev2
        prev2, prev1 = prev1, curr
    
    return prev1

# Test
print(climbStairs(1))  # 1
print(climbStairs(2))  # 2
print(climbStairs(3))  # 3
print(climbStairs(4))  # 5
print(climbStairs(5))  # 8

# Interview tip:
# - Recognize DP pattern: current depends on previous states
# - Recurrence: dp[i] = dp[i-1] + dp[i-2]
# - Space optimization: don't store all dp values if not needed
```

# 9. DECORATOR PATTERN

```python
"""Problem: Modify function behavior without changing source code
Example: Add logging, timing, caching, error handling
"""

import time
from functools import wraps

# DECORATOR 1: Timing
def timer(func):
    """Decorator that times function execution."""
    @wraps(func)  # Preserve func name and docstring
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        elapsed = time.time() - start
        print(f'{func.__name__} took {elapsed:.4f} seconds')
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(0.1)
    return 'Done'

# slow_function()  # Prints: slow_function took 0.1000 seconds

# DECORATOR 2: Retry Logic
def retry(max_attempts=3, delay=1):
    """Decorator that retries function on failure."""
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for attempt in range(1, max_attempts + 1):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    if attempt == max_attempts:
                        raise
                    print(f'Attempt {attempt} failed: {e}. Retrying in {delay}s...')
                    time.sleep(delay)
        return wrapper
    return decorator

@retry(max_attempts=3, delay=0.5)
def flaky_api():
    import random
    if random.random() < 0.7:
        raise ConnectionError('API unavailable')
    return 'Success'

# DECORATOR 3: Caching (LRU Cache)
from functools import lru_cache

@lru_cache(maxsize=128)
def expensive_computation(n):
    """Decorator caches function results."""
    print(f'Computing {n}...')
    return n ** 2

print(expensive_computation(5))  # Computing 5... Output: 25
print(expensive_computation(5))  # (cached) Output: 25

# DECORATOR 4: Logging
def log_calls(func):
    """Decorator that logs function calls."""
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f'Calling {func.__name__}({args}, {kwargs})')
        result = func(*args, **kwargs)
        print(f'{func.__name__} returned {result}')
        return result
    return wrapper

@log_calls
def add(a, b):
    return a + b

# add(2, 3)  # Prints: Calling add((2, 3), {})
            # add returned 5

# Interview tip:
# - @wraps preserves original function metadata
# - Decorators with arguments need double nesting
# - Common: retry, cache, logging, validation, authorization
```

# 10. CLASS & OOP BASICS

```python
# BASIC CLASS
class Person:
    """Basic class with init, methods, properties."""
    
    # Class variable (shared across instances)
    species = 'Homo sapiens'
    
    def __init__(self, name, age):
        """Constructor - called when creating instance."""
        self.name = name  # Instance variable
        self.age = age
    
    def greet(self):
        """Regular method - has access to self."""
        return f'Hello, I am {self.name}'
    
    @property
    def is_adult(self):
        """Property - accessed like attribute, not method."""
        return self.age >= 18
    
    @is_adult.setter
    def is_adult(self, value):
        """Setter - allows assignment to property."""
        if value:
            self.age = 18
        else:
            self.age = 10
    
    @staticmethod
    def is_valid_age(age):
        """Static method - no self/cls, utility function."""
        return 0 <= age <= 150
    
    @classmethod
    def from_birth_year(cls, name, birth_year):
        """Class method - receives cls, can create instance."""
        from datetime import datetime
        age = datetime.now().year - birth_year
        return cls(name, age)
    
    def __str__(self):
        """String representation - used by print()."""
        return f'{self.name} ({self.age} years old)'
    
    def __repr__(self):
        """Developer representation - used for debugging."""
        return f'Person({self.name!r}, {self.age})'

# Usage
p1 = Person('Alice', 25)
print(p1.greet())          # Hello, I am Alice
print(p1.is_adult)         # True (property access)
print(Person.is_valid_age(30))  # True (static method)

p2 = Person.from_birth_year('Bob', 2005)  # Class method
print(p2)  # Bob (19 years old)

# INHERITANCE
class Employee(Person):
    """Subclass inherits from Person."""
    
    def __init__(self, name, age, emp_id):
        super().__init__(name, age)  # Call parent constructor
        self.emp_id = emp_id
    
    def greet(self):  # Override parent method
        return f'{super().greet()} (ID: {self.emp_id})'

e1 = Employee('Charlie', 30, 'E001')
print(e1.greet())  # Hello, I am Charlie (ID: E001)

# Interview tip:
# - __init__: constructor
# - @property: attribute-like access to method
# - @staticmethod: utility function, no self
# - @classmethod: can access/modify class state
# - super(): access parent class methods
# - __str__ vs __repr__: str for users, repr for devs
```

# 11. GENERATORS & ITERATORS

```python
"""Generators: functions that yield values (lazy evaluation).
Useful for: large datasets, streaming, memory efficiency
"""

# GENERATOR FUNCTION
def countdown(n):
    """Generator - yields values one at a time."""
    while n > 0:
        yield n  # Pause here, save state
        n -= 1

# Usage
gen = countdown(3)
print(next(gen))  # 3
print(next(gen))  # 2
print(next(gen))  # 1
# print(next(gen))  # StopIteration

# Or iterate
for num in countdown(3):
    print(num)  # 3, 2, 1

# GENERATOR EXPRESSION (lazy list comprehension)
gen_expr = (x**2 for x in range(5))  # Generator expression
list_comp = [x**2 for x in range(5)]  # List comprehension

print(list(gen_expr))   # [0, 1, 4, 9, 16]
print(list_comp)        # [0, 1, 4, 9, 16]
# Generator is lazy: computes on-demand
# List: computes all at once

# CUSTOM ITERATOR
class Fibonacci:
    """Custom iterator class."""
    def __init__(self, max_count):
        self.max_count = max_count
        self.count = 0
        self.a, self.b = 0, 1
    
    def __iter__(self):
        """Return iterator object (usually self)."""
        return self
    
    def __next__(self):
        """Return next value. Raise StopIteration when done."""
        if self.count >= self.max_count:
            raise StopIteration
        
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

# Usage
fib = Fibonacci(5)
for num in fib:
    print(num)  # 0, 1, 1, 2, 3

# GENERATOR PIPELINE
def read_large_file(filepath):
    """Generator - read file line by line (memory efficient)."""
    with open(filepath) as f:
        for line in f:
            yield line.strip()

def filter_lines(lines, keyword):
    """Generator - filter lines containing keyword."""
    for line in lines:
        if keyword in line:
            yield line

def count_matches(lines):
    """Generator - count matching lines."""
    count = 0
    for line in lines:
        count += 1
        yield count

# Pipeline: read -> filter -> count
# Large files processed efficiently without loading all into memory

# Interview tip:
# - Generators are memory efficient (lazy evaluation)
# - yield pauses function, saves state
# - Iterators need __iter__() and __next__()
# - Generator expressions: (x for x in range(n)) vs [x for x in range(n)]
```

# 12. CONTEXT MANAGERS

```python
"""Context managers: resources cleanup guaranteed (__enter__ + __exit__).
Use case: file handling, database connections, locks, transactions
"""

# BUILT-IN: with statement
# with open('file.txt') as f:
#     content = f.read()
# # File auto-closes even if exception occurs

# CUSTOM CONTEXT MANAGER - Class approach
class DatabaseConnection:
    """Custom context manager for database."""
    
    def __init__(self, db_name):
        self.db_name = db_name
        self.connection = None
    
    def __enter__(self):
        """Called when entering with block."""
        print(f'Connecting to {self.db_name}...')
        self.connection = f'Connection to {self.db_name}'  # Mock
        return self.connection  # What 'as' variable gets
    
    def __exit__(self, exc_type, exc_val, exc_tb):
        """Called when exiting with block (even on exception)."""
        print(f'Closing connection to {self.db_name}...')
        self.connection = None
        
        # Return True to suppress exception
        if exc_type is not None:
            print(f'Exception occurred: {exc_val}')
        return False  # Don't suppress, let exception propagate

# Usage
with DatabaseConnection('postgres') as conn:
    print(f'Using: {conn}')
    # Connection automatically closed here

# CUSTOM CONTEXT MANAGER - Decorator approach (simpler)
from contextlib import contextmanager

@contextmanager
def database_connection(db_name):
    """Context manager using decorator."""
    print(f'Connecting to {db_name}...')
    conn = f'Connection to {db_name}'  # Setup
    try:
        yield conn  # What 'as' variable gets
    finally:
        print(f'Closing {db_name}...')  # Cleanup (guaranteed)

# Usage
with database_connection('mysql') as conn:
    print(f'Using: {conn}')
    # Connection auto-closes here

# CONTEXT MANAGER WITH EXCEPTION HANDLING
@contextmanager
def timer_context(name):
    """Context manager that times code execution."""
    import time
    start = time.time()
    print(f'Starting {name}...')
    try:
        yield
    finally:
        elapsed = time.time() - start
        print(f'{name} completed in {elapsed:.4f}s')

# Usage
with timer_context('computation'):
    total = sum(range(1000000))

# Interview tip:
# - __enter__: setup resources
# - __exit__: cleanup (called even on exception)
# - @contextmanager decorator is simpler for most cases
# - Use for: file ops, DB connections, locks, transactions
```

# 13. DEEP COPY VS SHALLOW COPY

```python
import copy

"""Deep copy vs shallow copy: critical difference!"""

# SHALLOW COPY: Copies object, nested objects are references
original = [[1, 2], [3, 4]]
shallow = original.copy()  # or copy.copy(original)

print('Original:', original)
print('Shallow:', shallow)
print('Same object?', original is shallow)  # False
print('Same nested object?', original[0] is shallow[0])  # True - PROBLEM!

# Modify nested object
shallow[0][0] = 99
print('After shallow[0][0] = 99:')
print('Original:', original)  # [[99, 2], [3, 4]] - AFFECTED!
print('Shallow:', shallow)   # [[99, 2], [3, 4]]

# DEEP COPY: Recursively copies all nested objects
original2 = [[1, 2], [3, 4]]
deep = copy.deepcopy(original2)

print('\nOriginal:', original2)
print('Deep:', deep)
print('Same nested object?', original2[0] is deep[0])  # False - Independent!

# Modify nested object
deep[0][0] = 99
print('After deep[0][0] = 99:')
print('Original:', original2)  # [[1, 2], [3, 4]] - NOT affected
print('Deep:', deep)          # [[99, 2], [3, 4]]

# COMPARISON TABLE
data = {'a': [1, 2, 3], 'b': [4, 5, 6]}
print('\n--- Comparison ---')
print('Original:', data)

# Assignment (reference)
ref = data
ref['a'][0] = 99
print('After ref["a"][0]=99:')
print('Original:', data)  # [99, 2, 3] - affected (same object)

# Shallow copy
data = {'a': [1, 2, 3], 'b': [4, 5, 6]}
shallow_copy = data.copy()
shallow_copy['a'][0] = 99
print('After shallow_copy["a"][0]=99:')
print('Original:', data)  # [99, 2, 3] - affected (nested ref)

# Deep copy
data = {'a': [1, 2, 3], 'b': [4, 5, 6]}
deep_copy = copy.deepcopy(data)
deep_copy['a'][0] = 99
print('After deep_copy["a"][0]=99:')
print('Original:', data)  # [1, 2, 3] - NOT affected

# Interview tip:
# - Assignment (=): reference, modifications affect both
# - Shallow copy (.copy()): copies object, nested refs unchanged
# - Deep copy (copy.deepcopy()): independent copy, modifications isolated
# - Know when to use each!
```

# 14. LAMBDA & MAP/FILTER/REDUCE

```python
"""Functional programming: lambda, map, filter, reduce"""

# LAMBDA: anonymous function
square = lambda x: x ** 2
print(square(5))  # 25

# MAP: apply function to each element
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print('map:', squared)  # [1, 4, 9, 16, 25]

# Equivalent list comprehension (preferred)
squared = [x**2 for x in numbers]
print('list comp:', squared)  # [1, 4, 9, 16, 25]

# FILTER: keep elements where condition is true
evens = list(filter(lambda x: x % 2 == 0, numbers))
print('filter:', evens)  # [2, 4]

# Equivalent list comprehension (preferred)
evens = [x for x in numbers if x % 2 == 0]
print('list comp:', evens)  # [2, 4]

# REDUCE: accumulate values
from functools import reduce

product = reduce(lambda x, y: x * y, numbers)
print('reduce (product):', product)  # 1*2*3*4*5 = 120

# Equivalent with explicit accumulation
product = 1
for x in numbers:
    product *= x
print('explicit:', product)  # 120

# Reduce with initial value
sum_val = reduce(lambda x, y: x + y, numbers, 100)
print('reduce (sum, init=100):', sum_val)  # 100+1+2+3+4+5 = 115

# CHAINING
data = [1, 2, 3, 4, 5]
# Filter evens, square them, sum
result = sum(x**2 for x in data if x % 2 == 0)
print('chained:', result)  # 2^2 + 4^2 = 4 + 16 = 20

# Interview tip:
# - Lambda: use for short, simple operations
# - Map/filter: less readable than list comprehensions
# - Prefer list comprehensions / generator expressions
# - Reduce: use for accumulation (sum, product, etc.)
```

# 15. SORTING & KEY FUNCTIONS

```python
"""Sorting: sorted(), list.sort(), custom key functions"""

# BASIC SORTING
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

# sorted(): returns new list, works on any iterable
sorted_nums = sorted(numbers)
print('sorted (ascending):', sorted_nums)  # [1, 1, 2, 3, 4, 5, 6, 9]

# reverse: descending
sorted_desc = sorted(numbers, reverse=True)
print('sorted (descending):', sorted_desc)  # [9, 6, 5, 4, 3, 2, 1, 1]

# list.sort(): in-place, only for lists
numbers.sort()
print('after sort():', numbers)  # [1, 1, 2, 3, 4, 5, 6, 9]

# CUSTOM KEY FUNCTION
words = ['apple', 'pie', 'a', 'python']

# Sort by length
by_length = sorted(words, key=len)
print('by length:', by_length)  # ['a', 'pie', 'apple', 'python']

# Sort by last character
by_last = sorted(words, key=lambda x: x[-1])
print('by last char:', by_last)  # ['apple', 'pie', 'python', 'a']

# SORTING OBJECTS
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
    def __repr__(self):
        return f'{self.name}({self.grade})'

students = [
    Student('Alice', 85),
    Student('Bob', 92),
    Student('Charlie', 78)
]

# Sort by grade (ascending)
by_grade = sorted(students, key=lambda s: s.grade)
print('by grade:', by_grade)  # Charlie(78), Alice(85), Bob(92)

# Sort by grade (descending)
by_grade_desc = sorted(students, key=lambda s: s.grade, reverse=True)
print('by grade desc:', by_grade_desc)  # Bob(92), Alice(85), Charlie(78)

# MULTI-KEY SORTING (tuple return)
data = [('a', 3), ('b', 1), ('a', 1), ('c', 2)]

# Sort by first, then second
sorted_data = sorted(data, key=lambda x: (x[0], x[1]))
print('multi-key:', sorted_data)  # [('a', 1), ('a', 3), ('b', 1), ('c', 2)]

# Interview tip:
# - sorted() returns new list, list.sort() modifies in-place
# - key parameter specifies sort criteria
# - Tuple keys sort by first element, then second, etc.
# - O(n log n) time complexity (Timsort)
```

# 16. MUTABLE VS IMMUTABLE & IS VS ==

```python
"""Understanding mutable vs immutable, and is vs =="""

# MUTABLE TYPES: can change in place
lst = [1, 2, 3]
lst[0] = 99  # Modify in place - OK
print('mutable list:', lst)  # [99, 2, 3]

# IMMUTABLE TYPES: cannot change in place, create new copy
s = 'hello'
# s[0] = 'H'  # ERROR - strings are immutable
s = 'H' + s[1:]  # Create new string
print('immutable string:', s)  # 'Hello'

# == vs is
print('\n--- == vs is ---')

# ==: checks VALUE equality
a = [1, 2, 3]
b = [1, 2, 3]
print('a == b:', a == b)  # True (same values)
print('a is b:', a is b)  # False (different objects)

# is: checks IDENTITY (same memory address)
c = a
print('a == c:', a == c)  # True (same values)
print('a is c:', a is c)  # True (same object)

# Small integers cached (-5 to 256)
x = 5
y = 5
print('x is y:', x is y)  # True (cached)

x = 257
y = 257
print('x is y:', x is y)  # False (not cached)

# Strings: interned in some cases
s1 = 'hello'
s2 = 'hello'
print('s1 is s2:', s1 is s2)  # True (usually interned)

s3 = 'hel' + 'lo'
print('s1 is s3:', s1 is s3)  # False (created dynamically)

# None check: use 'is'
value = None
if value is None:  # Correct
    print('value is None')

# NOT: if value == None (works but bad style)

# Interview tip:
# - Mutable: list, dict, set (can change)
# - Immutable: int, str, tuple, frozenset (cannot change)
# - == compares values
# - is compares identity (memory address)
# - Use 'is None', 'is True', 'is False' for checks
```

# 17. EXCEPTION HANDLING

```python
"""Exception handling: try-except-else-finally"""

# BASIC TRY-EXCEPT
try:
    result = 10 / 2
except ZeroDivisionError:
    print('Cannot divide by zero')
    result = 0

print('result:', result)  # 5.0

# MULTIPLE EXCEPT BLOCKS
def parse_int(value):
    try:
        return int(value)
    except ValueError:
        print(f'Cannot convert {value!r} to int')
        return 0
    except TypeError:
        print('Type error occurred')
        return 0

print(parse_int('42'))      # 42
print(parse_int('not_int')) # Cannot convert... 0

# CATCH MULTIPLE EXCEPTIONS
try:
    x = [1, 2, 3][5]  # IndexError
except (IndexError, KeyError) as e:
    print(f'Index or key error: {e}')

# ELSE & FINALLY
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print('Division by zero!')
    else:  # Runs if no exception
        print(f'{a} / {b} = {result}')
        return result
    finally:  # Always runs (cleanup)
        print('Division attempt completed')

safe_divide(10, 2)  # else runs
safe_divide(10, 0)  # except runs

# RAISE CUSTOM EXCEPTION
class InvalidAge(Exception):
    """Custom exception."""
    pass

def validate_age(age):
    if age < 0 or age > 150:
        raise InvalidAge(f'Invalid age: {age}')
    return age

try:
    validate_age(200)
except InvalidAge as e:
    print(f'Caught: {e}')

# EXCEPTION CHAINING
try:
    try:
        1 / 0
    except ZeroDivisionError as e:
        raise ValueError('Invalid calculation') from e
except ValueError as e:
    print(f'Final exception: {e}')
    print(f'Caused by: {e.__cause__}')

# Interview tip:
# - try: code that might fail
# - except: handle specific errors
# - else: runs if no exception
# - finally: always runs (cleanup, close files/connections)
# - raise: throw custom exceptions
```

# 19. MULTITHREADING & MULTIPROCESSING

```python
"""Multithreading vs Multiprocessing"""

import threading
import multiprocessing
import time

# MULTITHREADING (for I/O-bound tasks)
def io_task(task_id):
    """Simulates I/O task (network, file, etc.)."""
    print(f'Task {task_id} started')
    time.sleep(1)  # Simulate I/O wait
    print(f'Task {task_id} done')

# Create threads
threads = []
for i in range(3):
    t = threading.Thread(target=io_task, args=(i,))
    threads.append(t)
    t.start()  # Start thread

# Wait for all threads
for t in threads:
    t.join()  # Block until thread completes

print('All I/O tasks done')

# MULTIPROCESSING (for CPU-bound tasks)
def cpu_task(x):
    """CPU-intensive task."""
    total = 0
    for i in range(100000000):
        total += i
    return total

# Using Process Pool
if __name__ == '__main__':
    with multiprocessing.Pool(4) as pool:
        results = pool.map(cpu_task, range(4))
        print('CPU tasks done')

# Thread Pool (simpler)
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor

# For I/O tasks
with ThreadPoolExecutor(max_workers=3) as executor:
    futures = [executor.submit(io_task, i) for i in range(3)]
    # Wait for all to complete
    for future in futures:
        future.result()

# For CPU tasks
# with ProcessPoolExecutor(max_workers=4) as executor:
#     futures = [executor.submit(cpu_task, i) for i in range(4)]
#     results = [f.result() for f in futures]

# THREAD SYNCHRONIZATION
counter = 0
lock = threading.Lock()

def increment():
    global counter
    with lock:  # Acquire lock
        counter += 1  # Critical section
    # Lock released automatically

threads = [threading.Thread(target=increment) for _ in range(1000)]
for t in threads:
    t.start()
for t in threads:
    t.join()

print(f'Counter: {counter}')  # Should be 1000

# Interview tip:
# - Multithreading: I/O-bound (files, network, DB)
# - Multiprocessing: CPU-bound (math, ML training)
# - GIL: only 1 thread executes Python at a time (CPU bottleneck)
# - Use locks to prevent race conditions
# - ThreadPoolExecutor/ProcessPoolExecutor easier than manual threads
```

# 20. ASYNC/AWAIT (ADVANCED)

```python
"""Async programming: non-blocking, single-threaded concurrency"""

import asyncio
import time

# BASIC COROUTINE
async def fetch_data(url, delay):
    """Async function (coroutine)."""
    print(f'Fetching {url}...')
    await asyncio.sleep(delay)  # Await pauses, yields control
    print(f'Got {url}')
    return f'Data from {url}'

# RUN SINGLE COROUTINE
# result = asyncio.run(fetch_data('http://api.example.com', 1))
# print(result)

# RUN MULTIPLE CONCURRENTLY
async def main():
    """Run multiple coroutines concurrently."""
    # Method 1: gather
    results = await asyncio.gather(
        fetch_data('url1', 1),
        fetch_data('url2', 2),
        fetch_data('url3', 1)
    )
    print('All done:', results)

# asyncio.run(main())

# TIMEOUT
async def with_timeout():
    try:
        result = await asyncio.wait_for(
            fetch_data('slow', 5),
            timeout=2  # Timeout after 2 seconds
        )
    except asyncio.TimeoutError:
        print('Request timed out')

# asyncio.run(with_timeout())

# QUEUE
async def producer(queue):
    """Add items to queue."""
    for i in range(5):
        await asyncio.sleep(0.5)
        await queue.put(f'item-{i}')

async def consumer(queue):
    """Consume items from queue."""
    while True:
        item = await queue.get()
        print(f'Consumed: {item}')
        queue.task_done()
        if 'item-4' in item:
            break

async def queue_example():
    queue = asyncio.Queue()
    # Run both concurrently
    await asyncio.gather(
        producer(queue),
        consumer(queue)
    )

# asyncio.run(queue_example())

# Interview tip:
# - async def: coroutine function
# - await: pause, let other code run
# - asyncio.gather(): run multiple concurrently
# - Better than threading for I/O (no lock overhead)
# - Used in: web servers (FastAPI), scrapers, game loops
```
